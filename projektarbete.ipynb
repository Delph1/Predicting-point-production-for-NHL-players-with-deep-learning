```python
#snabb koll så säsongsdatan ser OK ut
df_2008 = pd.read_csv('skaters_2008.csv')
df_2008.head()
```

```python
#kikar på spelardatan också. Eftersom jag pillat med denna data och råkade spara den med semikolon istället för kolon som separator fick jag lägga till det som en parameter. 
df_players = pd.read_csv('players_edited.csv', sep=';')
df_players.head()
```

```python
#Vi skapar sedan en enda dataframe från alla säsonger och passar på att slå ihop den med spelarinformationen. Vi grupperar även informationen 

#2008 gjorde vi redan ovan
df_2009 = pd.read_csv('skaters_2009.csv')
df_2010 = pd.read_csv('skaters_2010.csv')
df_2011 = pd.read_csv('skaters_2011.csv')
df_2012 = pd.read_csv('skaters_2012.csv')
df_2013 = pd.read_csv('skaters_2013.csv')
df_2014 = pd.read_csv('skaters_2014.csv')
df_2015 = pd.read_csv('skaters_2015.csv')
df_2016 = pd.read_csv('skaters_2016.csv')
df_2017 = pd.read_csv('skaters_2017.csv')
df_2018 = pd.read_csv('skaters_2018.csv')
df_2019 = pd.read_csv('skaters_2019.csv')
df_2020 = pd.read_csv('skaters_2020.csv')
df_2021 = pd.read_csv('skaters_2021.csv')
df_2022 = pd.read_csv('skaters_2022.csv')
df_2023 = pd.read_csv('skaters_2023.csv')

df_all = pd.concat([df_2008, df_2009, df_2010, df_2011, df_2012, df_2013, df_2014, df_2015, df_2016, df_2017, df_2018, df_2019, df_2020, df_2021, df_2022, df_2023])

#Vi kollar hur stor vår dataframe blev. 
df_all.info()
```

```python
#Jag vill ha med spelarnas ålder som en parameter. Även om det bara blir en till av 150 och därmed troligen inte har så stor effekt i slutändan "känns" det bra att det är med. 
df_merged = df_all.merge(df_players[['playerId', 'birthDate']], left_on='playerId', right_on='playerId')

#Födelsedatum är dock inte helt enkelt att hantera så istället försöker vi konvertera det till ålder när säsongen börjar. 
#NHL brukar börja säsongen i oktober så jag sätter 1/10 som startdatum. Det är inte exakt, men bra nog. 

df_merged.insert(4, "age", "", True)

for index, row in df_merged.iterrows():
    birthDate = row['birthDate'].to_datetime
    
    if birthDate.month >= 10
        row['age'] = birthDate.year - row['season'] 
    else
        row['age'] = birthDate.year - row['season'] - 1
    print(row['birthDate'], row['age'])
    

#Vi behöver rensa bort en del rader också. 
#Målvakter kollar vi inte på nu.
df_merged = df_merged.loc[df_merged.position != 'G']

#Vi tar sedan bara ut situations = all
df_merged = df_merged.loc[df_merged.situation == 'all']

#Sedan grupperar vi datan per playerId och season
df_grouped = df_merged.groupby(['playerId', 'name', 'position', 'birthDate', 'season'])

#Snabb koll så att det ser OK ut:
df_grouped.head()
```
